### 常用创建数组
- zeros()
- ones()
- linspaces()

```python
# zeros: 创建一个全为0的数组
np.zeros(shape=(5,5),order='F') #形状, 内存储存风格
```

```python
# ones: 创建一个全为1的数组
np.ones(shape=(3,4))# 通常都是[行,列] 第一个参数为‘行’, 第二个参数为‘列 ’
```

```python
np.linspace(0, 100,num=20)#(start,stop, array number)
```

```python
np.arange(10, 50, step=2) # (开始,结束, 间隔)
```

### 索引和切片

```python
arr = np.random.randint(1, 100,size=(5,6))
arr
```

```python
arr[0,5] #[0行,5列]
```

```python
arr[0:2] #行切片:[0-2行]
```

```python
arr[:,0:2] #行列切片,切出arr数组的前两列
```

```python
arr[0:2,0:2] #切出前两行和前两列的数据
```

```python
arr[::-1]# 将行倒置
```

```python
arr[::-1,::-1]# 将所有元素倒置 [::]表示所有元素,可以指定步长
```

```python
imarr = imarr[:,::-1,:]
plt.imshow(imarr) # 行,列,颜色
```

```python
imarr = imarr[100:,::,:]
plt.imshow(imarr)
```

### 常用数学函数
- sin()、cos()、tan()
- around(x,decimals)函数返回指定数字的四舍五入
    - x 数组
    - decimals: 舍入的小数位数. 默认值为0. 如果为-, 整数将四舍五入到小数点左侧的位置

```python
np.cos(arr)
```

```python
np.around(3.14, 1)
```

### 常用统计函数
  - amin() and amax(), 用于计算数组中元素沿指定轴的最小,最大值.
  - ptp()计算最大值与最小值的差
  - median()计算数组a中元素的中位数(中值)
  - std() 标准差是一组数据平均值分散程度的一种度量
  - var() 统计中的方差 每个样本值与全体样本值的平均数之差的平方值的平均数

```python
np.ptp(arr,axis=1)
arr[0].ptp()
```
