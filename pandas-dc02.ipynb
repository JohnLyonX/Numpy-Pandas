### pandas处理空值操作
   - isnull
   - notnull
   - any
   - dropna
   - fillna

```python
import numpy as np
import pandas as pd
```

```python
# 伪造一组数据(存在控制)
df = pd.DataFrame(data=np.random.randint(0,100,size=(8,6)))

df.loc[2,3] = np.nan
df.loc[2,4] = np.nan
df.loc[3,1] = np.nan
df.loc[5,5] = None

df
```

- 方法1: 对空值进行过滤(删除空所在的行数据)
   - isnull 如果是空值的话返回True,否则返回Fales
   - notnull 如果不是空值的话返回True,否则返回Fales
   - any 如果有一个为真则返回True
   - dropna
   - fillna

```python
df.isnull() # 使用isnull
```

```python
ex = df.isnull().any(axis=1) # 对列进行查找,如果这个列有空值的话就返回True
df.loc[ex] # 这里已经看到有空值的数据了
df.drop(df.loc[ex].index) # index(返回所在行的索引)
```

```python
df.notnull() #使用notnull
```

```python
notex = df.notnull().all(axis=1) # axis=1 为纵轴,也就是从上到下进行操作 获取到所有有空值的列
df.loc[df.notnull().all(axis=1)]
```

- 搭配使用:
 - isnull -> any
 - notnull -> all

 - dropna: 可以直接将缺失的行或者列进行删除

```python
df.dropna(axis=0)
```

- 对缺失值进行覆盖
    - fillna

```python
df.fillna(value=0) # 将空值填充为0
df.fillna(method='ffill',axis=1) # 使用水平方向去向前填充空值
```
