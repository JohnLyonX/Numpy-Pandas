### 处理重复数据

```python
import numpy as np
# 生成一组带有重复数据的数据源(重复的行数据)
df = pd.DataFrame(data=np.random.randint(0,100,size=(8,4)))
df.iloc[2] = [0,0,0,0]
df.iloc[4] = [0,0,0,0]
df.iloc[5] = [0,0,0,0]
df.iloc[8] = [0,0,0,0]
```

```python
#使用drop_duplicates
df.drop_duplicates(keep='first') # first 保留第一个重复值 last保留最后一个重复值 False 删除所有重复的
```

 ### 处理异常值
   - 自定义一个1000行3列(A,B,C)取值范围为0-1的数据源,然后将C列中的值大于其两倍标准差的异常值进行清洗

```python
df = pd.DataFrame(data=np.random.random(size=(1000,3)),columns=['a','b','c'])
twice_std = df['c'].std() * 2

df['c'] > twice_std
(df['c'] > twice_std) 
-(df['c'] > twice_std) # -取反 True 变成 False
df.loc[-(df['c'] > twice_std)]
```
